## Chapter 2 - Plotting Time-Series

Many kinds of data are organized as time series and visualizations of time series are an excellent tool to detect patterns in the data. The co2 number is parts per million of carbon dioxide in the atmosphere on the given date. The relative_temp number is the temperature measured on this date relative to the baseline which is the average temperature in the first 10 years of measurements. To enable all of the capabilities of Pandas, when the data is imported, the keyword argument parse_date = will need to be provided a list of column(s) that should be recognized as date time series and the index_col = keyword argument also needs to be set to 'date'.

```python
import pandas as pd
climate_change = pd.read_csv('C:\\datacamp\\05-IntroductiontoDataVisualizationwithMatplotlib\\data\\climate_change.csv', parse_dates=['date'], index_col='date')
print(climate_change)
```

### Plotting Time-Series Data

To plot the data we create a figure and axes and then add the data to the plot create labels for the x and y axes. We see that matplotlib automatically chose to show the time as years with intervals of 10 years. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')
plt.show()
```

### Plotting Time-Series with Different Variables

We could add both of the variables from the climate_change DataFrame to a single plot, but the resulting plot doesn't tell a story since the scales for two measurements are different.

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'])
ax.plot(climate_change.index, climate_change['relative_temp'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm) / Relative temperature')
plt.show()
```

### Using Twin Axes

While using subplots can address the discrepancy between the two measurements' scales, there is also an option to have two different y axes using the .twinx() method to create a twin of this axis. This means that the two axes will share the same x-axis, but the y-axes are separate. We add the other variable to the second axis and then show the plot. 

#### Separating Variables by Color

By adding the color = keyword to both adding data to the plot and creating a label, the data becomes even more clear. 

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'])
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)')

ax2 = ax.twinx()

ax2.plot(climate_change.index, climate_change['relative_temp'])
ax2.set_ylabel('Relative temperature (Celsius)')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'], color='b')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='b')

ax2 = ax.twinx()

ax2.plot(climate_change.index, climate_change['relative_temp'], color='r')
ax2.set_ylabel('Relative temperature (Celsius)', color='r')
plt.show()
```

#### Color the Ticks

To further clarify the chart, an additional axes option can be set using the .tick_params() method. This method accepts an axis where ticks should be modified and the ability to change to color of the ticks. Note that the keyword argument for specifying the color is colors=, plural.

```python
fig, ax = plt.subplots()
ax.plot(climate_change.index, climate_change['co2'], color='b')
ax.set_xlabel('Time')
ax.set_ylabel('CO2 (ppm)', color='b')

ax.tick_params('y', colors='b')

ax2 = ax.twinx()

ax2.plot(climate_change.index, climate_change['relative_temp'], color='r')
ax2.set_ylabel('Relative temperature (Celsius)', color='r')
ax2.tick_params('y', colors='r')

plt.show()
```

### A Function That Plots Time-Series

A function that automatically pulls together all the work done above would facilitate creating and modifying time-series plots. This function takes the following arguments:
> an axes object<br>
> x and y variables to plot<br>
> a color to associate with the variable<br>
> x-axis and y-axis labels<br>

Using the function, we don't have to repeat these calls.

```python
def plot_timeseries(axes, x, y, color, xlabel, ylabel):
# A sample function for automating the creation of a time series plot

    axes.plot(x, y, color=color)
    axes.set_xlabel(xlabel)
    axes.set_ylabel(ylabel)
    axes.tick_params('y', colors=color)

fig, ax = plt.subplots()
plot_timeseries(ax, climate_change.index, climate_change['co2'], 'blue', 'Time', 'CO2 (ppm)')

ax2 = ax.twinx()
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], 'red', 'Time', 'Relative temperature (Celsius)')

plt.show()
```

#### Defining a function that plots time-series data
Once you realize that a particular section of code that you have written is useful, it is a good idea to define a function that saves that section of code for you, rather than copying it to other parts of your program where you would like to use this code.

Here, we will define a function that takes inputs such as a time variable and some other variable and plots them as x and y inputs. Then, it sets the labels on the x- and y-axis and sets the colors of the y-axis label, the y-axis ticks and the tick labels.

__Instructions:__
* Define a function called plot_timeseries that takes as input an Axes object (axes), data (x,y), a string with the name of a color and strings for x- and y-axis labels.
* Plot y as a function of in the color provided as the input color.
* Set the x- and y-axis labels using the provided input xlabel and ylabel, setting the y-axis label color using color.
* Set the y-axis tick parameters using the tick_params method of the Axes object, setting the colors key-word to color.

```python
# Define a function called plot_timeseries
def plot_timeseries(axes, x, y, color, xlabel, ylabel):

  # Plot the inputs x,y in the provided color
  axes.plot(x, y, color=color)

  # Set the x-axis label
  axes.set_xlabel(xlabel)

  # Set the y-axis label
  axes.set_ylabel(ylabel, color=color)

  # Set the colors tick params for y-axis
  axes.tick_params('y', colors=color)
```

### Annotating Time-Series Data

An important way to enhance visualizations is to add annotations. Annotations are small pieces of text that refer to part of the visualization; focusing attention a specific feature and explaining the feature. An example would be that the first date that the relative temperature exceeded 1 degree Celsius was October 6, 2015 and this could be pointed out on the plot. The .annote() method of the axes module and takes the annotation text as an input. It can also accept an xy coordinate to be annotated with the xy= keyword argument. 

```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change['co2'], "blue", 'Time (years)', 'CO2 levels')

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], "red", 'Time (years)', 'Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy=[pd.Timestamp('2015-10-06'), 1])

plt.show()
```

### Positioning the Text

This doesn't quite look right, as the annotation text is crossing over the y axis. Using the keyword argument xytext=, the position of the annotation can be defined. With some testing, it has been determined that October 6, 2008 is a good x value for the annotation and a y value of -0.2 is the best location for the annotation text.

```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change['co2'], "blue", 'Time (years)', 'CO2 levels')

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], "red", 'Time (years)', 'Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy=[pd.Timestamp('2015-10-06'), 1], xytext=(pd.Timestamp('2008-10-06'), -0.2))

plt.show()
```

#### Adding Arrows to Annotations

While the annotation is now readable and not covering other points on the chart, it is unclear what this annotation is referring to. We can add an arrow using the arrowprops= keyword argument. The arrowprops keyword takes a dictionary that defines the properties of the arrow we would like to use. Passing an empty dictionary will result in the arrow having the default properties.

```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change['co2'], "blue", 'Time (years)', 'CO2 levels')

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], "red", 'Time (years)', 'Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy=[pd.Timestamp('2015-10-06'), 1], xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={})

plt.show()
```

#### Customing Arrow Properties

The properties can be customized with keyword arguments like 'arrowstyle': and 'color':. Using the -> will create an arrow with a thin line and wide head. The customization options for annotations can be found at https://matplotlib.org/users/annotations.html

```python
fig, ax = plt.subplots()

# Plot the CO2 levels time-series in blue
plot_timeseries(ax, climate_change.index, climate_change['co2'], "blue", 'Time (years)', 'CO2 levels')

# Create a twin Axes object that shares the x-axis
ax2 = ax.twinx()

# Plot the relative temperature data in red
plot_timeseries(ax2, climate_change.index, climate_change['relative_temp'], "red", 'Time (years)', 'Relative temperature (Celsius)')

ax2.annotate('>1 degree', xy=[pd.Timestamp('2015-10-06'), 1], xytext=(pd.Timestamp('2008-10-06'), -0.2), arrowprops={'arrowstyle':'->', 'color':'gray'})

plt.show()
```
