## Chapter 1 - Introduction to Matplotlib

Matplotlib is a Python library widely used to visualize data. One of the advantages of Matplotlib is the complete control over the properties of the plot. The main object oriented interface of Matplotlib is pyplot. When the pyplot submodule is imported, the standard convention is to import as plt. plt.subplots() command, when called without any instance, creates two different objects, a figure object and an axes objects. The figure object is a container that hold everything visible on the page. The axes is the part of the page that holds the data. It is the canvas on which to draw the data for visualization. 

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
plt.show()
```

### Customizing Your Plots

When you look at the plot of Seattle's monthly percipitation as a line chart, the data appears to be continuous, but in fact, there are 12 points of data, one for each month. 

#### Adding Markers

Adding markers to indicate where the data points are helps to clarify the difference between the data and the line used to connect the data points. Setting the keyword argument markers = to lower case o uses filled in circles as markers. Setting the argument to lower case v will use filled in triangles. All the available marker styles are defined at https://matplotlib.org/api/markers_api.html

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='o')
plt.show()
```

#### Setting the Linestyle

With the markers in place, it is easier to see that the lines between the markers are just there to connect the data point. To further emphasize this, the line itself can be customized with the keyword argument, linestyle =. Using two dashes, --, will create a dashed line between the data points. Available linestyles can be found at https://matplotlib.org/gallery/lines_bars_and_markers/line_styles_reference.html. 

#### Eliminate lines with Linestyle
Passing the keyword "None" to linestyle will remove the line completely and only leave the datapoints.

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='v', linestyle='--')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='v', linestyle='None')
plt.show()
```

#### Choosing Color

The line and marker color can also be customized using the keyword argument color=. Setting the color argument to 'r' will change the color to red.

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='v', linestyle='--', color='r')
plt.show()
```

#### Customizing the Axes Labels

In addition to the .plot() method, the axes object has several methods that start with the word set_ can help customize the chart. .set_xlabel() will set a label for the x axis and it's corresponding .set_ylabel() to set a label for the y axis. The punctuation for labels is the same as sentences in that you only capitalize the first word in the label and proper nouns.

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='v', linestyle='--', color='r')
ax.set_xlabel("Time (months)")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='v', linestyle='--', color='r')
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
plt.show()
```

#### Adding a Title

The .set_title() method will place a title at the top of the graph.

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], marker='v', linestyle='--', color='r')
ax.set_xlabel("Time (months)")
ax.set_ylabel("Average temperature (Fahrenheit degrees)")
ax.set_title("Weather in Seattle")
plt.show()
```

### Small Multiples

In some cases, adding more data to plot can make it too busy, obscuring patterns rather than revealing them. If we start with the Seattle average rainfall per month and then add in the 25th and 75th percentiles rain fall as well, the 75th percentile will be the dashed line above the average and the 25th percentile will be the dashed line below the average. But, if we then add the some data for Austin, the resulting graph is a bit of a mess. 

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'b')
ax.plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'b')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax.plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'b')
ax.plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'b')

ax.plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])
ax.plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'r')
ax.plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'r')
plt.show()
```

One way to address this is to use small multiples. These are multiple small plots that show similar data across different conditions. For example, precipitation data across different cities. The code below creates a figure object with 3 rows of subplots and 2 columns. #### Subplots with Data

There is a special case where you only have 1 row or 1 column. This case you only have provide 1 element to access the index of this array. In the case of the rainfall for each city, we add the Seattle data to the subplot element 0 and the Austin data to the subplot element 1. 

We add y-axis labels for both plots, but because they are stacked on top of each other, only the bottom plot needs an x-axis label.

The last thing to be addressed is the range of the y-axes. Because the range of the data are not the same, the range of the y-axis, by default, are going to be dependent on the range of the data provided each subplot. When defining the subplot shape, using the keyword argument sharey= set to True will make sure all the subplots have the same range of y-axis values. 

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax[0].plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'b')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])
ax[1].plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'r')
ax[1].plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'r')
```

```python
fig, ax = plt.subplots(2,1)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax[0].plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'b')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])
ax[1].plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'r')
ax[1].plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'r')

ax[0].set_ylabel('Precipitation (inches)')
ax[1].set_ylabel('Precipitation (inches)')
ax[1].set_xlabel('Time (months)')
```

```python
fig, ax = plt.subplots(2,1, sharey=True)
ax[0].plot(seattle_weather["MONTH"], seattle_weather["MLY-PRCP-NORMAL"])
ax[0].plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'b')
ax[0].plot(seattle_weather["MONTH"], seattle_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'b')

ax[1].plot(austin_weather["MONTH"], austin_weather["MLY-PRCP-NORMAL"])
ax[1].plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-25PCTL'], linestyle = '--', color = 'r')
ax[1].plot(austin_weather["MONTH"], austin_weather['MLY-PRCP-75PCTL'], linestyle = '--', color = 'r')

ax[0].set_ylabel('Precipitation (inches)')
ax[1].set_ylabel('Precipitation (inches)')
ax[1].set_xlabel('Time (months)')
```

#### Creating small multiples with plt.subplots
Small multiples are used to plot several datasets side-by-side. In Matplotlib, small multiples can be created using the plt.subplots() function. The first argument is the number of rows in the array of Axes objects generate and the second argument is the number of columns. In this exercise, you will use the Austin and Seattle data to practice creating and populating an array of subplots.

The data is given to you in DataFrames: seattle_weather and austin_weather. These each have a MONTH column and MLY-PRCP-NORMAL (for average precipitation), as well as MLY-TAVG-NORMAL (for average temperature) columns. In this exercise, you will plot in a separate subplot the monthly average precipitation and average temperatures in each city.

__Instructions:__
* Create a Figure and an array of subplots with 2 rows and 2 columns.
* Addressing the top left Axes as index 0, 0, plot the Seattle precipitation.
* In the top right (index 0,1), plot Seattle temperatures.
* In the bottom left (1, 0) and bottom right (1, 1) plot Austin precipitations and temperatures.

```python
# Create a Figure and an array of subplots with 2 rows and 2 columns
fig, ax = plt.subplots(2, 2)

# Addressing the top left Axes as index 0, 0, plot month and Seattle precipitation
ax[0,0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'])

# In the top right (index 0,1), plot month and Seattle temperatures
ax[0,1].plot(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'])

# In the bottom left (1, 0) plot month and Austin precipitations
ax[1,0].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'])

# In the bottom right (1, 1) plot month and Austin temperatures
ax[1,1].plot(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'])
plt.show()
```

#### Small multiples with shared y axis
When creating small multiples, it is often preferable to make sure that the different plots are displayed with the same scale used on the y-axis. This can be configured by setting the sharey key-word to True.

In this exercise, you will create a Figure with two Axes objects that share their y-axis. As before, the data is provided in seattle_weather and austin_weather DataFrames.

__Instructions:__
* Create a Figure with an array of two Axes objects that share their y-axis range.
* Plot Seattle's "MLY-PRCP-NORMAL" in a solid blue line in the top Axes.
* Add Seattle's "MLY-PRCP-25PCTL" and "MLY-PRCP-75PCTL" in dashed blue lines to the top Axes.
* Plot Austin's "MLY-PRCP-NORMAL" in a solid red line in the top Axes and the "MLY-PRCP-25PCTL" and "MLY-PRCP-75PCTL" in dashed red lines.

```python
# Create a figure and an array of axes: 2 rows, 1 column with shared y axis
fig, ax = plt.subplots(2, 1, sharey=True)

# Plot Seattle precipitation data in the top axes
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-NORMAL'], color = 'b')
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-25PCTL'], color = 'b', linestyle = '--')
ax[0].plot(seattle_weather['MONTH'], seattle_weather['MLY-PRCP-75PCTL'], color = 'b', linestyle = '--')

# Plot Austin precipitation data in the bottom axes
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-NORMAL'], color = 'r')
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-25PCTL'], color = 'r', linestyle = '--')
ax[1].plot(austin_weather['MONTH'], austin_weather['MLY-PRCP-75PCTL'], color = 'r', linestyle = '--')

plt.show()
```
