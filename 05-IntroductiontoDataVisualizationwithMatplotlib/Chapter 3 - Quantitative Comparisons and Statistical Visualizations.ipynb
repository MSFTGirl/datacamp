This chapter will be focused on quantitative comparisons between parts of the data. The following dataset represents the number of medals won by a few countries from the 2016 Olympics. ### Olympic Medals: Visualizing the Data

Using pandas, we read the file and select the first column, the country, as the index for the data and then we visualize the data for gold medals. The labels of the x-axis correspond to the index of the DataFrame which contains the country names. The y-axis the height of a bar that represents the number of gold medals each country was awarded.

```python
import pandas as pd
import matplotlib.pyplot as plt
medals = pd.read_csv('C:\\datacamp\\05-IntroductiontoDataVisualizationwithMatplotlib\\data\\medals.csv', index_col=0)
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
plt.show()
```

#### Interlude: Rotate the Tick Labels

 Because the names of the x-axis are rather long, they overlap each other and that be addressed with the .set_xticklabels() method and set the rotation = keyword argument to 90 to rotate the text by 90 degrees. We can also add a label for the y-axis to indicate the taht height corresponds to the number of medals won.

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of gold medals won')
plt.show()
```

### Olympic Medals: Visualizing the Other Medals

We can also add the data about the other medals to this bar chart. To add the data regarding the number of silver and bronze medals each country was awarded, we create a stacked bar chart, where the data is stacked on top of each other using the bottom = argument to tell matplotlib that data specified in this argument should be at the bottom of the height of the previous data.

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'])
ax.set_xticklabels(medals.index, rotation = 90)
ax.set_ylabel("Number of Silver and Gold Medals")
plt.show()
```

#### Olympic Medals: Visualizing All Three

Similarly, we can add the number of bronze medals to the graph by setting the bottom of this bar to the sum of number of gold and silver medals. 

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'])
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'])
ax.bar(medals.index, medals['Bronze'], bottom = medals['Gold'] + medals['Silver'])
ax.set_xticklabels(medals.index, rotation = 90)
ax.set_ylabel("Number of Bronze, Silver and Gold Medals")
plt.show()
```

#### Adding a Legend

To make the figure easier to read and understand, a legend can be added identifying what the colors of each bar represents. To do this we need to add the label = keyword argument to each call of the .bar() method and then add an ax.legend() method before call the .show() method

```python
fig, ax = plt.subplots()
ax.bar(medals.index, medals['Gold'], label = 'Gold')
ax.bar(medals.index, medals['Silver'], bottom = medals['Gold'], label = 'Silver')
ax.bar(medals.index, medals['Bronze'], bottom = medals['Gold'] + medals['Silver'], label = 'Bronze')
ax.set_xticklabels(medals.index, rotation = 90)
ax.set_ylabel("Number of Medals")
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.bar('Rowing', mensrowing['Height'].mean())
ax.bar('Gymnastics', mensgymnastics['Height'].mean())
ax.set_ylabel('Height (cm)')
plt.show()
```

### Introducing Histograms

A histogram would show the full distribution of values within each variable. In the histogram below, the x-axis is the values within the variable and the height of the bar represents the number of observations with a particular bin of values. Because there is no information provided about which color represents which sport, labels are needed in histograms.

```python
fig, ax = plt.subplots()
ax.hist(mensrowing['Height'])
ax.hist(mensgymnastics['Height'])
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
plt.show()
```

```python
fig, ax = plt.subplots()
ax.hist(mensrowing['Height'], label='Rowing')
ax.hist(mensgymnastics['Height'], label='Gymnastics')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

#### Customizing Histograms: Setting the Number of Bins

By default, the number of bins used by a histogram is 10. This can be customized by providing an integer number to the keyword argument bin= when defining the histogram.

```python
fig, ax = plt.subplots()
ax.hist(mensrowing['Height'], label='Rowing', bins=5)
ax.hist(mensgymnastics['Height'], label='Gymnastics', bins=5)
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

#### Customizing Histograms: Setting Bin Boundaries

Providing a sequence of numbers to the bins argument, these numbers will be set to be the boundaries between the bins.

```python
fig, ax = plt.subplots()
ax.hist(mensrowing['Height'], label='Rowing', bins=[150,160,170,180,190,200,210])
ax.hist(mensgymnastics['Height'], label='Gymnastics', bins=[150,160,170,180,190,200,210])
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

#### Customizing Histograms: Transparency

Looking at the above figure, it is not possible to determine if there are any rowing medalist with a height of less that 180 cm as the bars for the gymnastics histogram are occluding this information. The occlusion can be eliminated by changing the type of histogram used. Instead of a bar type that is used by default, a step type can be used by setting the keyword argument histtype = to 'step' which displays the histogram as thin lines rather than solid bars. 

```python
fig, ax = plt.subplots()
ax.hist(mensrowing['Height'], label='Rowing', bins=[150,160,170,180,190,200,210],histtype='step')
ax.hist(mensgymnastics['Height'], label='Gymnastics', bins=[150,160,170,180,190,200,210],histtype='step')
ax.set_xlabel('Height (cm)')
ax.set_ylabel('# of observations')
ax.legend()
plt.show()
```

### Statistical Plotting

Statistical plotting is a set of methods for using visualization to make comparisons. 

#### Adding Error Bars to Bar Charts

Error bars in plots are additional markers on a plot or bar chart that visualizes the distribution of the data. Histograms in the previous activites show the entire distribution; error bars summarize the distribution in one number. In the example below the error bar is added as an argument when creating the bar chart. The yerr= keyword argument takes an additional number, in this case the standard deviation of the height measurement, and displays this as an additional vertical marker. 

With the error bar added, the plot summarizes the full distribution in the original histogram in two numbers: the mean value and the spread value quantified as the standard deviation. 

```python
fig, ax = plt.subplots()
ax.bar('Rowing', mensrowing['Height'].mean(), yerr=mensrowing['Height'].std())

ax.bar('Gymnastics', mensgymnastics['Height'].mean(), yerr=mensgymnastics['Height'].std())

ax.set_ylabel('Height (cm)')

plt.show()
```

#### Adding Error Bars to Plots

We can also add error bars to line charts. Using the weather data from Seattle and Austin and the .errorbar() method on the axes. Like the .plot() method, the .errorbar() method takes a sequence of x values (the Month column for this chart), a sequence of y values (the Monthly Average Temperature for this chart) and, in addition, a yerr= keyword argument which is assigned to the Standard Deviation of the Monthly Average Temperature column. With the .errorbar() method, vertical markers are added to the plot.

```python
fig, ax = plt.subplots()

ax.errorbar(seattle_weather['MONTH'], seattle_weather['MLY-TAVG-NORMAL'], yerr=seattle_weather['MLY-TAVG-STDDEV'])
ax.errorbar(austin_weather['MONTH'], austin_weather['MLY-TAVG-NORMAL'], yerr=austin_weather['MLY-TAVG-STDDEV'])
ax.set_ylabel('Temperature (Fahrenheit)')
plt.show()
```

### Adding Boxplots

Another statistical visualization is the boxplot, invented by John Tucci, the first data scientist. Using the axes .boxplot() method and providing a sequence of sequences; in this case a list with the height columns from the men's rowing and gymnastics data. Adding labels is necessary to understand the data visualization. 

The resulting plot shows several landmarks in each distribution:
* Red line is the median height.
* The edges of the box portion at the center indicate the interquartile range of the data, between the 25th and 75th percentiles of the data. 
* The whiskers at the end of the thin bar indicate 1.5 times the size of the interquartile range beyond the 75th and 25th percentiles. This should encompass roughly 99% of the distribution.
* Points that appear beyond the whiskers are outliers; values that are larger or smaller than what you would expect for 99% of the data.

```python
fig, ax = plt.subplots()

ax.boxplot([mensrowing['Height'], mensgymnastics['Height']])
ax.set_xticklabels(['Rowing', 'Gymnastics'])
ax.set_ylabel('Height (cm)')

plt.show()
```

### Introducing Scatter Plots

Bar charts show the values of one variable across different conditions. To compare the value of two different variables across observations, a bivariate comparison, a standard visualization is a scatter plot. Using the climate change data, each measurement for CO2 and Temperature can be represented as a point with the distance along the x-axis representing the measurement in column and the height on the y-axis representing the measurement in the other column.

Using the .scatter() method, the first argument will correspond to the measurement along the x-axis and the second argument will correspond to the height along the y-axis. 

```python
fig, ax = plt.subplots()

ax.scatter(climate_change['co2'], climate_change['relative_temp'])
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
plt.show()
```

#### Customing Scatter Plots

Scatter plots can be customized in a similar way that the other plot types are. When more than one bi-varient comparisons are being shown on a scatter plot, they will need to be visually distinct. In the example below, the data from the 1980s is customized by setting the keyword argument color = to red following adding the 80s dataset to the plot. The same is done for the 90s dataset, setting the color= argument to blue.

```python
eighties = climate_change['1980-01-01':'1989-12-31']
nineties = climate_change['1990-01-01':'1999-12-31']

fig, ax = plt.subplots()
ax.scatter(eighties['co2'], eighties['relative_temp'], color='red', label='eighties' )
ax.scatter(nineties['co2'], nineties['relative_temp'], color='blue', label='nineties')
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
ax.legend()
plt.show()
```

#### Encoding A Third Variable by Color

The color of the points can be used to encode a third variable to provide additional information about the comparisons. The continuous variable of time is stored in the DataFrame index. If we add the index to the keyword argument c =, the date will get encoded as color. In the resulting chart, the time of the measurement is indicated in the brightness of the point's color with the dark blue point early on and the bright yellow for the later points. 

```python
fig, ax = plt.subplots()

ax.scatter(climate_change['co2'], climate_change['relative_temp'], c=climate_change.index)
ax.set_xlabel('CO2 (ppm)')
ax.set_ylabel('Relative temperature (Celsius)')
plt.show()
```
