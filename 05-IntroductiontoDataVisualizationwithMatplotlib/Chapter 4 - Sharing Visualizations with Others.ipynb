#### Guidelines for Choosing a Plot Style

If the goal is to primarily communicate with others, think about how the audience will interpret the chart. 
* Dark backgrounds are generally discouraged as they are usually less visible
* If colors are important, consider using a color-blind friendly styles such as seaborn-colorblind or tableau-colorblind10
* If the plot is going to be printed, avoid solid colored elements to use less ink
* If the printer will likely be black and white, consider using the grayscale style

#### Different File Formats

The .png file format provides lossless compression of the image which means the image will retain high quality, but also be large in size. Using the .jpg file format will use lossy compression and can be used to create figures that are going to take up less disk space and less bandwidth. Controlling the size of the file and the degree to the loss of quality using the quality = keyword argument when saving the plot.

Choosing a .svg file format will produce a vector graphics file. This means that different elements of the plot can be edited in detail by advanced graphics software. If the figure is going to need to be edited, this should be the file format selected. 

#### Resolution

The dpi = keyword argument will control the quality of the file format as well. Dots Per Inch, dpi, will determine the density of the image; the higher the number, the more densely the image will be rendered. Setting the dpi value to 300 will render a fairly high quality image file which will result in a larger file size as well. 

#### Size

Another characteristic that can be controlled is the size of the image using the .set_size_inches() function, providing it a list of numbers that represent the width and height and therefore defining the aspect ratio. 

```python
plt.style.use('default')
fig, ax = plt.subplots()

ax.bar(medals.index, medals['Gold'])
ax.set_xticklabels(medals.index, rotation=90)
ax.set_ylabel('Number of medals')
fig.set_size_inches([3,5])

plt.show()
```

### Automating Figures from Data

Matplotlib can flexible adapt to the inputs that are provided. This means that functions and programs can be developed that automatically adjust what they are doing based on the input data. Inspecting incoming data and changing the behavior of the program provides robustness and reproducibility. 

#### Getting Unique Values of a Column

Using the summer 2016 medal data on all of the sports, calling the .unique method on the Sports column in the all medals DataFrame will return a list of all the sports that made up the 2016 summer Olympics. To visualize the height of the athletes in all of these sports with the standard deviation error bar, we first extract the unique values of the sports that make up the dataset and then loop over each value. Then calling the .bar() method of the axes, the elements of the graph can be set, in this case, the mean of the height and an error bar set to the standard deviation of the height of each athlete by sport. 

Once all of the sports have been looped over to create the axes, the y label can be set the x tick label can be defined. At no point did the number of sports have to be defined or known in order to create the resulting graph, it adapted to the data provided. 

```python
sports = summer_2016_medals['Sport'].unique()
print(sports)

fig, ax = plt.subplots()

for sport in sports:
    sport_df = summer_2016_medals[summer_2016_medals['Sport'] == sport]
    ax.bar(sport, sport_df['Height'].mean(), yerr=sport_df['Height'].std())

ax.set_ylabel('Height (cm)')
ax.set_xticklabels(sports, rotation=90)

plt.show()
```

#### Automate your visualization
One of the main strengths of Matplotlib is that it can be automated to adapt to the data that it receives as input. For example, if you receive data that has an unknown number of categories, you can still create a bar plot that has bars for each category.

This is what you will do in this exercise. You will be visualizing data about medal winners in the 2016 summer Olympic Games again, but this time you will have a dataset that has some unknown number of branches of sports in it. This will be loaded into memory as a Pandas DataFrame object called summer_2016_medals, which has a column called "Sport" that tells you to which branch of sport each row corresponds. There is also a "Weight" column that tells you the weight of each athlete.

__Instructions:__
* Iterate over the values of sports setting sport as your loop variable.
* In each iteration, extract the rows where the "Sport" column is equal to sport.
* Add a bar to the provided ax object, labeled with the sport name, with the mean of the "Weight" column as its height, and the standard deviation as a y-axis error bar.
* Save the figure into the file "sports_weights.png".

```python
fig, ax = plt.subplots()

# Loop over the different sports branches
for sport in sports:
  # Extract the rows only for this sport
  sport_df = summer_2016_medals[summer_2016_medals['Sport']==sport]
  # Add a bar for the "Weight" mean with std y error bar
  ax.bar(sport, sport_df['Weight'].mean(), yerr=sport_df['Weight'].std())

ax.set_ylabel("Weight")
ax.set_xticklabels(sports, rotation=90)

# Save the figure to file
fig.savefig('sports_weights.png')
```
