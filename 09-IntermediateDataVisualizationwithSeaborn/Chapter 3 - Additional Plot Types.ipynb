### Categorical Plots

Categorical data has a limited and fixed number of values and is usually combined with numeric data. Seaborn breaks categorical data plots into 3 groups:

#### Plots of each observation
Plots that show all the individual observations, such as stripplots and swarmplots are often difficult to read due to the concentration of each observation. 

Stripplots have a jitter = keyword argument that when set to True makes it easier to see each observation point.


The swarmplot has a complex algorithm that better shows each point of observation by ensuring the do not overlap. This means that the swarmplot does not scale well to large datasets. 


#### Plots of abstract representations
Plots that show abstract representation of the data with plots such as boxplots, violinplots and LVplots. 

The boxplot is the most common, showing several measures related to the distribution of the data, including the median, upper and lower quartiles, as well as outliers. 


The violin plot is a combination of a kernel density plot and a boxplot and provides an alternative view of the distribution of the data. Because the violinplot uses a kernel density plot, it does not show all data points and this can be useful for large datasets, but is often computationally intensive to create.


The lvplot, which stands for letter value plot, is a hybrid between the boxplot and the violinplot and can scale more effectively to large datasets. 


#### Statistical estimates
Plots that show statistical estimates of the categorical variables, such as barplots, pointplots and countplots.

The barplot shows the estimate of the value as well as a confidence interval.


The pointplot is similar to the boxplot in that it shows a summary measure and confidence interval. The point plot is useful for observing how values change across categorical values. 


The countplot displays the number of instances of each variable

### Regression Plots

Seaborn supports sophisticated regession plot capabilities, as this type of plot is important in data science. Like other plots, the regression plot takes a data source and columns for the x and y axes. The marker = keyword argument allows for definition of a specific character to be used for each data point. 


#### Residual Plots

Residual plots are useful for understanding the appropriateness of a regression model. Ideally, the residual values in a plot should be plotted randomly across the horizontal line. In this example, the data appears to have a slight curve, suggesting that a non-linear model might be appropriate.


#### Polynomial Regression

If a value greater than 1 is passed to the order = parameter of regplot, Seaborn will attempt a polynomial fit using underlying NumPy functions. In this example, Seaborn computes a second order polynomial function for the relationship between temperature and rentals. 


#### Residual Plts with Polynomial Regression

By adding the second order, it shows that the data is more randomly distributed, indicating that a second order is more appropriate for this specific data. 


#### Categorical Plots

Seaborn also supports regression plots with categorical variables. In this example, using the x_jitter = paramter makes it easier to see the rental values for each month.


##### Estimators

Sometimes, even with using the jitter = parameter, it is difficult to see the values in a categorical regression plot. Using the x_estimator = paramter and setting it to a NumPy summary function like np.mean. This will help to visualize trends in the data.


##### Binning the Data
When there are continuous variables, it may be helpful to break them into bins. By setting the x_bins = parameter to a specific number, Seaborn will take care of calculating the bins and plotting the results.

Linear regression is a useful tool for understanding the relationship between numerical variables. Seaborn has simple but powerful tools for examining these relationships.

For these exercises, we will look at some details from the US Department of Education on 4 year college tuition information and see if there are any interesting insights into which variables might help predict tuition costs.

__Instructions:__

```python
# Load dataset used in exercise
df = pd.read_csv('2018_college_data.csv')

# Display a regression plot for Tuition
sns.regplot(data=df,
         y='Tuition',
         x="SAT_AVG_ALL",
         marker='^',
         color='g')

plt.show()
plt.clf()
```

### Matrix Plots

The heatmap is the most common type of matrix plot and is easily created by Seaborn and allows to quickly see trends in data. Seaborn's heatmap function expects the data to be in matrix. The data can be transformed to a matrix format using the Pandas crosstab() function. Passing the DataFrame column slices for the x and y axes and setting the values = parameter a column slice and the aggfunc = parameter an aggregration type, a crosstab of the data can be created.


The heatmap uses the crosstab to translate the values into a color coded grid. In this example, the higher numbers are reflected with lighter colors, showing that Saturdays in May and June are the busiest times for bike rentals.


#### Customing Heatmaps

Customization allows the displaying of the most information in the clearest way possible. The annot = parameter turns on annotations in the individual cells. The format = option ensures that the results are displayed as integers. The cmap = option changes the color scheme and the cbar = option determines if the color bar is displayed (True) or not displayed (False). the linewidth = option defines the size of the lines that should be included between each cell. 


The center = variable accepts an .loc[] location within the crosstab. In this example the locations for September (9) and Saturday (6) are provided and the effect is that the overall color scheme is shifted to yellow.


#### Plotting a Correlation Matrix

One common usage of a heatmap is to visually represent the correlation between variables. Pandas has a .corr() function that can be applied to a DataFrame to calculate the correlation between the values in the columns. The output of this structure is ideally displayed as a heatmap. This visualization helps to identify which variables might require further studying using regplot. This plot shows that the rentals are most highly correlated with the temp variables.

#### Creating heatmaps
A heatmap is a common matrix plot that can be used to graphically summarize the relationship between two variables. For this exercise, we will start by looking at guests of the Daily Show from 1999 - 2015 and see how the occupations of the guests have changed over time.

The data includes the date of each guest appearance as well as their occupation. For the first exercise, we need to get the data into the right format for Seaborn's heatmap function to correctly plot the data. All of the data has already been read into the df variable.

__Instructions:__
* Use pandas' crosstab() function to build a table of visits by Group and Year.
* Print the pd_crosstab DataFrame.
* Plot the data using Seaborn's heatmap().

```python
# Load data used for these exercises
df = pd.read_csv('daily_show_guests.csv')

# Create a crosstab table of the data
pd_crosstab = pd.crosstab(df["Group"], df["YEAR"])
print(pd_crosstab)

# Plot a heatmap of the table
sns.heatmap(pd_crosstab)

# Rotate tick marks for visibility
plt.yticks(rotation=0)
plt.xticks(rotation=90)

plt.show()
```

#### Customizing heatmaps
Seaborn supports several types of additional customizations to improve the output of a heatmap. For this exercise, we will continue to use the Daily Show data that is stored in the df variable but we will customize the output.
