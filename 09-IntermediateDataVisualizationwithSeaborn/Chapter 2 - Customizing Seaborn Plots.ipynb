# Intermediate Data Visualization with Seaborn

## Chapter 2 - Customizing Seaborn Plots

### Using Seaborn Styles

Choosing the right type plot to effectively display the data is only one aspect; other aspects include layouts, labels and colors. There are several default Seaborn styles that are more visually appealing than the default matplotlib styles. The default Seaborn style can be set as the global default for plots by calling the .set() function. By calling sns.set(), the defaut styles of both Pandas and matplotlib are overwritten by the Seaborn style default, darkgrid. The colors are muted and a grey background with grid lines are applied to the plot. 



### Theme Examples with sns.set_style()

Styles control multiple aspects of a visualization, including the color palette and the use of grid lines and tick marks. The code below shows an example of the different default styles. One style may be able to better clarify the visualization over another and the best approach is to experiment with the style types. 

```python
for style in ['white', 'dark', 'whitegrid', 'darkgrid','ticks']:
    sns.set_style(style)
    sns.distplot(df_college['Tuition'])
    plt.show()
```

#### Removing Axes with .despine()

Sometimes plot understanding can be clarified by removing elements from the plot. Often the removal of the y-axis or x-axis line results in a cleaner plot and that can be accomplished using the Seaborn .despine() function. Setting the left= keyword argument to True will result in a plot without the y-axis line. 

```python
sns.set_style('white')
sns.distplot(df_college['Tuition'])
sns.despine(left=True)
```

#### Removing spines

In general, visualizations should minimize extraneous markings so that the data speaks for itself. Seaborn allows you to remove the lines on the top, bottom, left and right axis, which are often called spines.

__Instructions:__
* Use a white style for the plot.
* Create a lmplot() comparing the pop2010 and the fmr_2 columns.
* Remove the top and right spines using despine().

```python
# Set the style to white
sns.set_style('white')

# Create a regression plot
sns.lmplot(data=df_fmr,
           x='pop2010',
           y='fmr_2')

# Remove the spines
sns.despine(left=True, top=True)

# Show the plot and clear the figure
plt.show()
plt.clf()
```
