### Sorting and Subsetting

Two of the simplest and most important ways to find interesting parts of your data. The first thing is changing the order of rows by sorting by column variable values using the .sort_values() method and passing in a column name that you want to sort by. Setting the ascending argument to False will sort the data from heaviest to lightest. 

```python
print(dogs.sort_values('weight_kg'))
print()
print(dogs.sort_values('weight_kg', ascending = False))
```

#### Sorting by Multiple Variables

Multiple variables can be used to sort the data by passing a list of column names to sort by, in the order the columns are listed in the list. 

To change the direction of how the data is sorted, pass the ascending argument a list of True or False statements that corresponds to the list of columns.

```python
print(dogs.sort_values(['weight_kg', 'height_cm']))
print()
print(dogs.sort_values(['weight_kg', 'height_cm'], ascending = [True, False]))
```

#### Subsetting Columns

When it is necessary to zoom in on a column, the column name is placed in square brackets [] following the name of the DataFrame itself. 

```python
dogs['Name']
```

#### Subsetting Multiple Columns

To select multiple columns, provide a list of column names to the subsetting square brackets following the name of the DataFrame itself. By using two sets of square brackets, the name of a list variable that is created separately can be passed to the subsetting square brackets. 

```python
dogs[["breed", "height_cm"]]
```

```python
cols_to_subset = ['breed', 'height_cm']
dogs[cols_to_subset]
```

#### Subsetting Rows

Typically a row is subset by creating a logical condition to filter against.

```python
dogs['height_cm'] > 50
```

By using the logical condition inside of subsetting square brackets, the rows will be subsetted to those that meet the logical condition. 

```python
dogs[dogs['height_cm']>50]
```

#### Subsetting Based on Text Data

Rows can also be subset using text data using the double equal size. 

```python
dogs[dogs['breed'] == 'Labrador']
```

#### Subsetting Based on Dates

Date columns can also be used to subset the data.

```python
dogs[dogs['dob']> "2015-01-01"]
```

#### Subsetting Based on Multiple Conditions

Condition statements can be combined using logical operators, such as the AND operator represented by an ampersand. This can also be done with one line of code, using parenthesis to separate each condition and the logical operator to join the conditions.

```python
is_lab = dogs['breed'] == 'Labrador'
is_brown = dogs['color'] == 'Brown'
dogs[is_lab & is_brown]
```

```python
dogs[(dogs['breed']=="Labrador") & (dogs['color']=='Brown')]
```

#### Subsetting using .isin()

If the sort needs to be based categorical variables, the easiest way is to use the .isin() method. Passing a list of categorical variables to the .isin() method will subset the rows that have the same values as defined on the list passed to the .isin() method.

```python
is_black_or_brown = dogs['color'].isin(["Black", "Brown"])
dogs[is_black_or_brown]
```

### New Columns

There may be times when new columns will be needed, often derived from data in existing columns. Adding new columns can be referred to as mutating the DataFrame, transforming the DataFrame and "feature engineering". To add a column to the dogs DataFrame that stores the dogs' height in meters, on the left of the equal size you type the DataFrame that you want to modify followed by the name of the to be created column within square brackets and quotation marks. On the right of the equal sign is the calculation that should be used to populate the fields of the new column.

```python
dogs['height_m'] = dogs['height_cm']/100
print(dogs)
```

#### Doggy Mass Index

To calculate and store the dogs' Body Mass Index (BMI), the weight in KG is divided by the height in M squared: weight in kg/(height in m)^2.

```python
dogs['bmi'] = dogs['weight_kg'] / dogs['height_m']**2
print(dogs)
```

### Mutiple Manipulations

Finally, only display the columns that are related to the search: dog name, height_cm and BMI.

```python
bmi_lt_100 = dogs[dogs['bmi']<100]
bmi_lt_100_height = bmi_lt_100.sort_values('height_cm', ascending = False)
bmi_lt_100_height[['Name', 'height_cm', 'bmi']]
```

#### Combo-attack!
You've seen the four most common types of data manipulation: sorting rows, subsetting columns, subsetting rows, and adding new columns. In a real-life data analysis, you can mix and match these four manipulations to answer a multitude of questions.<br>
<br>
In this exercise, you'll answer the question, "Which state has the highest number of homeless individuals per 10,000 people in the state?" Combine your new pandas skills to find out.<br>

__Instructions:__
*  Add a column to homelessness, indiv_per_10k, containing the number of homeless individuals per ten thousand people in each state.
*  Subset rows where indiv_per_10k is higher than 20, assigning to high_homelessness.
*  Sort high_homelessness by descending indiv_per_10k, assigning to high_homelessness_srt.
*  Select only the state and indiv_per_10k columns of high_homelessness_srt and save as result. Look at the result.

```python
# Create indiv_per_10k col as homeless individuals per 10k state pop
homelessness['indiv_per_10k'] = 10000 * homelessness['individuals'] / homelessness['state_pop']

# Subset rows for indiv_per_10k greater than 20
high_homelessness = homelessness[homelessness['indiv_per_10k'] > 20]

# Sort high_homelessness by descending indiv_per_10k
high_homelessness_srt = high_homelessness.sort_values('indiv_per_10k', ascending = False)

# From high_homelessness_srt, select the state and indiv_per_10k cols
result = high_homelessness_srt[['state', 'indiv_per_10k']]

# See the result
print(result)
```
