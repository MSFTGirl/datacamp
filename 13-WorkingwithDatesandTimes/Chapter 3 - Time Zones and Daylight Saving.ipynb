### UTC Offsets

Naive timeset are not able to be compared across different parts of the world because there is no time zone attached. With UTC being set in England, time zones west of England are minus UTC and those east of England are plus UTC. The timezone class from the datetime module and is imported like datetime and timedelta. Time zone objects are created using a time delta that explains how to translate the datetime time zone from UTC. United States Eastern Time is UTC - 5, so a -5 timedelta is used to create the Eastern Time timezone. The timezone object can then be added to the tzinfor= argument when creating datetime objects. Now when the dt datetime object is printed, it will include the UTC offset information as well. 

```python
from datetime import datetime, timedelta, timezone
EST = timezone(timedelta(hours=-5))
dt = datetime(2017,12,30,15,9,3, tzinfo=EST)
print(dt)
```

Datetime objects have an .astimezone() method that accepts a timezone object as an argument and returns what the corresponding time is for the time zone specified.

```python
IST = timezone(timedelta(hours = 5, minutes = 30))
print(dt.astimezone(IST))
```

#### Adjusting Time Zones vs Changing the tzinfo Directly

There is a difference between adjusting time zone and changing the tzinfo = argument directly. Using the .replace() method, the tzinfo can be updated. Notice that both times are the same, the only change is the UTC offset, negative 5 for dt originally and 0 when the tzinfo was set to UTC. 

When the .astimezone() method is used, the actual time is adjusted to be the exact same moment, but with the time of the passed timezone. 

```python
print(dt)
print(dt.replace(tzinfo = timezone.utc))
print(dt.astimezone(timezone.utc))
```

```python
# Import datetime, timezone
from datetime import datetime, timezone

# October 1, 2017 at 15:26:26, UTC
dt = datetime(2017, 10, 1, 15, 26, 26, tzinfo=timezone.utc)

# Print results
print(dt.isoformat())

# Create a timezone for Pacific Standard Time, or UTC-8
pst = timezone(timedelta(hours=-8))

# October 1, 2017 at 15:26:26, UTC-8
dt = datetime(2017, 10, 1, 15, 26, 26, tzinfo=pst)

# Print results
print(dt.isoformat())

# Create a timezone for Australian Eastern Daylight Time, or UTC+11
aedt = timezone(timedelta(hours=+11))

# October 1, 2017 at 15:26:26, UTC+11
dt = datetime(2017, 10, 1, 15, 26, 26, tzinfo=aedt)

# Print results
print(dt.isoformat())
```

### Time Zone Database

There are more than 38 different time zones in use around the world. Rather than having to look up each time zone and its offset from UTC, Python has a time zone database called tz that is updated 3 or 4 times a year as time zone rules change. The tz database is available in the dateutil package. Within tz, time zones are defined first by the continent it is on followed by a forward slash and then the nearest major city. By calling tz.gettz and passing 'America/New_York' a timezone object is created for the time zone that corresponds to the eastern seaboard of the United States. Other examples include:

'America/Mexico_City'
'Europe/London'
'Africa/Accra'

Looking at the last ride from the onebike data, rather than creating the timezone object direct, the time zone obtained from tz is passed to tzinfo= when creating the datetime object. Looking at the results, it shows it has the correct UTC offset. 

The advantage of using the tz database to create the timezone objects is that it automatically accounts for the start and stop of daylight savings. Using the et time zone object when creating a datetime object for the bike's first ride, creates an object with a UTC offset of -4 to accomodate still being a daylight savings time. The tz database has time zone rules that go as far back as the late 1960s. 

```python
from datetime import datetime
from dateutil import tz

#Eastern Timezone
et = tz.gettz('America/New_York')

last = datetime(2017, 12, 30, 15, 9, 3, tzinfo=et)
print(last)

first = datetime(2017, 10, 1, 15, 23, 25, tzinfo=et)
print(first)
```

```python
# Import datetime, timedelta, tz, timezone
from datetime import datetime, timedelta, timezone
from dateutil import tz

# Start on March 12, 2017, midnight, then add 6 hours
start = datetime(2017, 3, 12, tzinfo = tz.gettz('America/New_York'))
end = start + timedelta(hours=6)
print(start.isoformat() + " to " + end.isoformat())


# How many hours have elapsed?
print((end - start).total_seconds()/(60*60))

# What if we move to UTC?
print((end.astimezone(timezone.utc) - start.astimezone(timezone.utc))\
      .total_seconds()/(60*60))
```

### Ending Daylight Saving Time

On November 5, 2017 at 2 am in Washington, DC the clocks jump back an hour, that means there were 2 1 ams on the same day. This is represented by folding over the timeline to show the repeat.



As before, to make sense of this, it needs to be mapped back to UTC. The first 1 am maps to 5 am UTC. This is a -4 hours UTC offset for Eastern Daylight Time. Once the clock jumps backwards, it is necessary to take the time from UTC. We switch to a UTC offset of -5 hours colored in blue and the second 1 am corresponds to 6 am UTC.



To demonstrate, first create a timezone for eastern time using tz.gettz(). The create a datetime object for November 5, 1:00 am. Then use the tz.datetime_ambiguous() method. If it comes back as True, it means that this time could occur at 2 different UTC moments in this time zone. 

```python
from datetime import datetime
eastern = tz.gettz('US/Eastern')
first_1am = datetime(2017,11,5,1,0,0, tzinfo=eastern)
tz.datetime_ambiguous(first_1am)
```

Now we create a second datetime, using the same date and time. This time we call tz.enfold() which takes the argument of a datetime. Enfold says this datetime belongs to the second time the wall clock struck 1 am on this day and not the first. To be clear, .enfold() does not change any behavior of a datetime, subtracting first_1am from second_1am is still going to give a result of 0. Fold is just a placeholder and it's up to other parts of the program to pay attention this. Once the times are converted to UTC, 

```python
import datetime
eastern = tz.gettz('US/Eastern')
second_1am = datetime(2017,11,5,1,0,0, tzinfo=eastern)
second_1am = tz.enfold(second_1am)

print(first_1am - second_1am.total_seconds())

first_1am = first_1am.astimezone(timezone.utc)
second_1am = second_1am.astimezone(timezone.utc)

print(first_1am - second_1am.total_seconds())
```
