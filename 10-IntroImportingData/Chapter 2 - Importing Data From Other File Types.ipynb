### Introduction to Other File Types

Other files types that data scientists use including Excel files, SAS files, MATLAB, Stata and HDF5 file formats. Another file type is pickled files, a native to Python file type. 

#### Pickle Files
Some Python objects like Pandas DataFrames and NumPy arrays are easily saved to flat files, but other objects like dictionaries and lists are not, and such the need for the pickle file type. If the file needs to be human readable, than exporting text files in a format like json is an option, but, if all you need to do is import the data stored in dictionaries or lists, you can use pickle to serialize the data. Serialization is the process of converting the object into a sequences of bytes or a bytestream. When you open a pickled file, you will want to use the r argument to specify it is read only, but also use the b argument to specify that is a binary file and therefore not human readable.

```python
#Behind the scenes data recreation
import os
os.chdir('c:\\datacamp\\data\\')
import pickle
data = {'peaches': '13', 'apples': '4', 'oranges': '11'}
pickle_out = open('pickled_fruit.pkl', 'wb')
pickle.dump(data, pickle_out)
```

```python
# Import pickle package
import pickle
 
with open('pickled_fruit.plt', 'rb') as file:
   data = pickle.load(file)
print(data)
```

#### Importing sheets from Excel files
In the previous exercises, you saw that the Excel file contains two sheets, '2002' and '2004'. The next step is to import these.

In this exercise, you'll learn how to import any given sheet of your loaded .xlsx file as a DataFrame. You'll be able to do so by specifying either the sheet's name or its index.

The spreadsheet 'battledeath.xlsx' is already loaded as xls.

__Instructions:__

* Print the head of df1 to the shell.
* Load the sheet 2002 into the DataFrame df2 using its index (0).

```python
# Load a sheet into a DataFrame by name: df1
df1 = xls.parse('2004')

# Print the head of the DataFrame df1
print(df1.head())

# Load a sheet into a DataFrame by index: df2
df2 = xls.parse(0)

# Print the head of the DataFrame df2
print(df2.head())
```

#### Customizing your spreadsheet import
Here, you'll parse your spreadsheets and use additional arguments to skip rows, rename columns and select only particular columns.

The spreadsheet 'battledeath.xlsx' is already loaded as xls.

As before, you'll use the method parse(). This time, however, you'll add the additional arguments skiprows, names and usecols. These skip rows, name the columns and designate which columns to parse, respectively. All these arguments can be assigned to lists containing the specific row numbers, strings and column numbers, as appropriate.

__Instructions:__ 
* Parse the first sheet by index. In doing so, skip the first row of data and name the columns 'Country' and 'AAM due to War (2002)' using the argument names. The values passed to skiprows and names all need to be of type list.
* Parse the second sheet by index. In doing so, parse only the first column with the usecols parameter, skip the first row and rename the column 'Country'. The argument passed to usecols also needs to be of type list.

```python
# Parse the first sheet and rename the columns: df1
df1 = xls.parse(0, skiprows=1, names=('Country',"AAM due to War (2002)"))

# Print the head of the DataFrame df1
print(df1.head())

# Parse the first column of the second sheet and rename the column: df2
df2 = xls.parse(1, usecols=[0], skiprows=[0], names=['Country'])

# Print the head of the DataFrame df2
print(df2.head())
```

### Importing HDF5 Files

With the Python community, there is growing support for the Hierarchical Data Format version 5, or HDF5, file type for storing large quantities of numerical data. These datasets can be hundreds of gigabytes or even terabytes in size. HDF5 can scale up to exabytes. To use HDF5 files, you need to import the pacakge h5py and then use the .File() function to open the file, using  the 'r' parameter to specify read-only. 

```python
import h5py
filename = 'LIGO.hdf5'
data = h5py.File(filename, 'r')
print(type(data))
```

#### The Structue of an HDF5 File

You can explore the hierarchical structure of an HDF5 file much like you do a Python dictionary object, using the .keys() method. Each of the keys is an HDF group, and the groups are similar to directories. The LIGO documentation tells us that meta contains metadata for the file, quality contains information about data quality and strain contains strain data from the interferometer, the primary measurement performed by LIGO. 




From within the meta group, you can also print the keys and further explore the data within the HDF5 file. Once you have the keys, you can access the data within each key. 

```python
print('Groups in the HDF5 File:')
for key in data.keys():
    print(key)

print('\r')
print('Python object type of the meta group:' + str(type(data['meta'])))
print('\r')

print('Keys in the meta group:')
for key in data['meta'].keys():
    print(key)

print('\r')
print(data['meta']['Description'].value, data['meta']['Detector'].value)
```

#### Extracting data from your HDF5 file
In this exercise, you'll extract some of the LIGO experiment's actual data from the HDF5 file and you'll visualize it.

To do so, you'll need to first explore the HDF5 group 'strain'.

__Instructions:__
* Assign the HDF5 group data['strain'] to group.
* In the for loop, print out the keys of the HDF5 group in group.
* Assign to the variable strain the values of the time series data data['strain']['Strain'] using the attribute .value.
* Set num_samples equal to 10000, the number of time points we wish to sample.
* Execute the rest of the code to produce a plot of the time series data in LIGO_data.hdf5.

```python
import numpy as np
import matplotlib.pyplot as plt

# Get the HDF5 group: group
group = data['strain']

# Check out keys of group
for key in group.keys():
    print(key)

# Set variable equal to time series data: strain
strain = data['strain']['Strain'].value

# Set number of time points to sample: num_samples
num_samples = 10000

# Set time vector
time = np.arange(0, 1, 1/num_samples)

# Plot data
plt.plot(time, strain[:num_samples])
plt.xlabel('GPS Time (s)')
plt.ylabel('strain')
plt.show()
```

### Loading MATLAB Files

MATLAB is short for Matrix Laboratory is a numerical computing environment and an industry standard in engineering and science. MATLAB files use the .mat file extension. MATLAB files can be imported using the SciPy package and the .io.loadmat() function to read .mat files and .io.savemat() function to write .mat files.

#### What is a .mat file?
In the MATLAB IDE, you see that the MATLAB workspace is where all the variables are stored that can be strings, floats, vectors and arrays and many other object and the the .mat file is a simply a collection of these objects.



When you use scipy.io to load a .mat file, it creates a Python dictionary object and the keys in the dictionary are the MATLAB variable names and the values are the objects assigned to the variable.

```python
import scipy.io
filename = 'agexpress.mat'
mat = scipy.io.loadmat(filename)
print(type(mat))
```
