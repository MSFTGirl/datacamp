### Histograms

Histograms are useful visualizations when you are trying to get an idea of how your data is distributed. For example:
Imagine 12 values from 0 to 6, as shown in the number line below:



To build a histogram for these values, you can divide the line into equal chunks or bins. Supposed you go with 3 bins and each have a weight of 2. 



You then count how many data points sit inside each bin. There are 4 data points within the first bin, 6 in the second bin and 2 in the last bin.



Finally you draw a bar for each bin and the height of the bar corresponds to the number of data points in that bin. Providing an overview of how the 12 values are distributed.



```python
help(plt.hist)
```

You can see there are a number of arguments for the histogram, the first two are the most important. X is a list values to build the histogram with and bins is the number that the data should be divided.
