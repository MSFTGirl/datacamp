### Introducton to Relational Plots and Subplots

Relational plots look at the relationship between two quantitative variables: do taller people weigh more (height vs weight), do the number of school absences impact a student's grade (number of absences vs final grade) or do countries with larger GDPs have higher literacy rates (GDP vs literacy rate).

When it is possible that relational data may be different in different subgroups, it may be helpful to not just use the hue = keyword argument, but to plot the relationship between two quantitative variables using either scatter plots or line plots, but as subplots in a single figure using the Seaborn function .relplot(). Similar to using the .scatterplot() function, the .relplot() function takes column name or list for the x and y axes, requires the data source to be defined in the data = keyword argument, but also takes a kind = keyword argument that will specify the type of plot the .relplot() function should create. By setting the col = keyword argument to the 'smoker' column, .replot() knows to create two plots, one in each column, for smokers (yes) and smoker (no). 

```python
import seaborn as sns
import matplotlib.pyplot as plt

tips = sns.load_dataset('tips')

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker')
plt.show()
```

#### Subplots in Rows

If it would be preferred to have the subplots stacked on top of each other rather than side by side, use the row = keyword argument in place of the col = keyword argument.

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', row='smoker')
plt.show()
```

#### Subplots in Rows and Columns

There are situations where it makes sense to use both the row = and col = keyword arguments. Below, row = is set to the 'time' column in the tips dataset and col = is set to the 'smoker' column in the tips dataset, resulting in 4 subplots.

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='smoker', row='time')
plt.show()
```

### Subgroups for Days of the Week

When looking at the total ball vs the tip amount across the days of the week, by default, 4 subplots will be displayed in single row, which can be a lot. Using the col_wrap = keyword argument will define how many columns of plots should be displayed per row. The col_order = keyword can also be used to provide a list of the ordered values to show the subplots.

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='day')
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='day', col_wrap=2)
plt.show()
```

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', col='day', col_wrap = 2,col_order=['Thur','Fri','Sat','Sun'])
plt.show()
```

### Customizing Scatter Plots

Scatter plots are great for visualizing the relationship between 2 quantitative variables. Additional customizations that Seaborn offers includes customizing subgroup point size and style, as well as point transparency. 

#### Subgroups with Point Size

In creating the tips vs bill scatter plot, we can change the size of the point using the size = keyword argument and setting it to 'size' based on the number of people in the group. Point size is best used when the variable is a quantitative variable or a categorical variable that represents different levels. 

#### Point Size and Hue
The plot is difficult to read since all the points are of the same color, but using the hue = parameter as well, and setting it to 'size' will add clarification to the plot. Notice that Seaborn colors the points in different shades of the same color since these data points are quantitative variables, rather than in different colors like it did for categorical data in the smoker vs non-smoker plots. 

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', size='size')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', size='size', hue='size')
```

#### Subgroups with Point Styles

In addition to color and size, the point style can be customized using the style = keyword argument. Setting this parameter to another variable name, Seaborn will use different point styles for each value of the variable. 

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', hue='smoker', style = 'smoker')
```

#### Changing Point Transparency

Using the alpha = keyword argument and setting the parameter to a value between 0 and 1, will vary the transparency of the point on the plot with 0 being completely transparent and 1 being completed nontransparent. This customization can be useful when there are many overlapping points on the scatter plot and it is possible to see which areas of the plot have more or less observations. 

```python
sns.relplot(x='total_bill', y='tip', data=tips, kind='scatter', alpha=.4)
```

### Introduction to Line Plots

Line plots are the visualization used to track the same variable over time. Creating a line plot is similar to creating a scatter plot, providing Seaborn's .relplot() function with a variable for the x axis and one for the y axis and specifying 'line' for the kind = parameter. Looking at weather data, we can show the average temperature by month in a line plot. The resulting plot makes it easy to see how the average monthly temperature flucuates throughout the year. 

```python
sns.relplot(x='MONTH', y='MLY-TAVG-NORMAL', data=seattle, kind = 'line')
plt.show()
```

#### Subgroups By Location

Subgroups can also tracked over time with line plots. Using a combined weather dataset of weather statistics from two cities, Seattle and Austin, setting the style = and hue = keywords to the 'NAME' column will create different lines for each city's monthly average temperature over the year. 

```python
sns.relplot(x='MONTH', y='MLY-TAVG-NORMAL', data=weather, kind = 'line', style='NAME', hue='NAME')
plt.show()
```

#### Adding Markers

Setting the markers = parameter to True will add a marker for each data point. By default the line and marker will by diffent for each value in the column set in the style= parameter. If you don't want the lines styles to vary by subgroup, set the dashes = parameter to false.

```python
sns.relplot(x='MONTH', y='MLY-TAVG-NORMAL', data=weather, kind = 'line', style='NAME', hue='NAME', markers = True)
plt.show()
```

```python
sns.relplot(x='MONTH', y='MLY-TAVG-NORMAL', data=weather, kind = 'line', style='NAME', hue='NAME', markers = True, dashes=False)
plt.show()
```

#### Multiple Observations Per x-Value

Line plots can also be used when you have more than observation per value. In the case of the weather dataset of both Austin and Seattle monthly weather measurements, there will be two observations for every monthly variable, one for Seattle and one for Austin. When a line plot is given multiple observations per x value, it will aggregate them into a single summary measure and by default will display the mean. Seaborn will also provide a confidence interval for the mean displayed in the shaded area. 

```python
sns.relplot(x='MONTH', y='MLY-TAVG-NORMAL', data=weather, kind = 'line')
plt.show()
```

#### Replacing Confidence Interval with Standard Deviation

If visualizing the variance between the two x-axis variable is preferred over the confidence interval, use the ci = keyword argument and set it to 'sd' for standard deviation which shows the spread of observations of each x value. 

By setting the ci = parameter to None, it will disable the confidence interval.

```python
sns.relplot(x='MONTH', y='MLY-TAVG-NORMAL', data=weather, kind = 'line', ci='sd')
```
