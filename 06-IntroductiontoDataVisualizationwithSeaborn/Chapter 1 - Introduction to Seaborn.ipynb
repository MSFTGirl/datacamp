# Introduction to Data Visualization with Seaborn

## Chapter 1 - Introduction to Seaborn

Seaborn is a Python data visualization library developed to create the most common types of plots. Seaborn was first developed to handle a lot of complexity behind the scenes, making data visualization easy and works well with pandas data structures. Seaborn is typically imported as sns which stands for Samuel Norman Seaborn, the package's namesake and the character from The West Wing TV show. In addition, because Seaborn is built on top of matplotlib, it too must be installed in order to use Seaborn.

### Example 1: Scatter Plot

For the first Seaborn visualization example, explore if taller people weigh more by plotting the data below into a scatter plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

height = [62,64,69,75,66,68,65,71,76,73]
weight = [120,136,148,175,137,165,154,172,200,187]

sns.scatterplot(x=height, y=weight)
plt.show()
```

### Example 2: Count Plot

A count plot takes a categorical list and return bars that represent the number of list entries per category. By calling the countplot() function and passing it the list of genders, we see the visualization of the gender count of male and female.

```python
gender = ['Female', 'Female', 'Female', 'Female', 'Male', 'Male', 'Male', 'Male','Male','Male']

sns.countplot(x=gender)
```

### Adding a Third Variable with Hue

#### Tips Dataset

Seaborn as a built in dataset named tips. Using the Seaborn .load_dataset() function and passing it the name 'tips' will automatically load the built in dataset. A scatterplot can be used to explore the relationship between the total bill amount and the tip amount. 

```python
tips = sns.load_dataset('tips')
print(tips.head())

sns.scatterplot(x='total_bill', y='tip', data=tips)
plt.show()
```

#### Adding a Hue

To see which of the tips were left by smokers versus non-smokers, use the keyword argument hue = and set it the smoker column. Note that Seaborn automatically created a legend to clarify which tips were smokers and which were non-smokers.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker')
plt.show()
```

#### Setting Hue Order

In addition to the keyword hue =, the keyword argument hue_order can be passed a list to specify the order of the items listed in the legend.

```python
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', hue_order=['Yes', 'No'])
plt.show()
```

#### Specifying Hue Colors

By creating a Python dictionary of responses and the desired color for each response and passing the dictionary name to the keyword argument palette = .

Hue is a keyword argument for many types of Seaborn charts.

```python
hue_colors = {'Yes':'black', 'No':'red'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', hue_order=['Yes', 'No'], palette=hue_colors)
plt.show()
```

#### List of Colors From Matplotlib

This is a list of available colors and their hexcodes and abbreviations that can be used to customize the Seaborn hue colors:



```python
hue_colors = {'Yes':'#808080', 'No':'#00FF00'}
sns.scatterplot(x='total_bill', y='tip', data=tips, hue='smoker', hue_order=['Yes', 'No'], palette=hue_colors)
plt.show()
```

In the prior video, we learned how hue allows us to easily make subgroups within Seaborn plots. Let's try it out by exploring data from students in secondary school. We have a lot of information about each student like their age, where they live, their study habits and their extracurricular activities.

For now, we'll look at the relationship between the number of absences they have in school and their final grade in the course, segmented by where the student lives (rural vs. urban area).

#### Hue and count plots
Let's continue exploring our dataset from students in secondary school by looking at a new variable. The "school" column indicates the initials of which school the student attended - either "GP" or "MS".

In the last exercise, we created a scatter plot where the plot points were colored based on whether the student lived in an urban or rural area. How many students live in urban vs. rural areas, and does this vary based on what school the student attends? Let's make a count plot with subgroups to find out.

__Instructions:__
* Fill in the palette_colors dictionary to map the "Rural" location value to the color "green" and the "Urban" location value to the color "blue".
* Create a count plot with "school" on the x-axis using the student_data DataFrame.
* Add subgroups to the plot using "location" variable and use the palette_colors dictionary to make the location subgroups green and blue.

```python
# Import Matplotlib and Seaborn
import matplotlib.pyplot as plt
import seaborn as sns

# Create a dictionary mapping subgroup values to colors
palette_colors = {'Rural': "green", 'Urban': "blue"}

# Create a count plot of school with location subgroups
sns.countplot(x='school', data=student_data, hue='location', palette=palette_colors)



# Display plot
plt.show()
```
