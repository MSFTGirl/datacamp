### Changing Plot Style and Color

There are 5 figure styles that include background and axes customizations:
> white<br>
> dark<br>
> whitegrid<br>
> darkgrid<br>
> ticks<br>

Using the Seaborn .set_style() function and passing it one of the built in style names will set the selected style as the global style type. 

#### Default Figure Style - white

The default style is white. This is a useful style when looking to provide a comparison between groups. When looking to show a general trend across age groups without specific values, the default style is a good choice.

```python
sns.catplot(x='age', y='masculinity_important', data=masculinity_data, hue = 'feel_masculine', kind='point', order=['18 - 34', '35 - 64', '65 and up'])
plt.show()
```

#### Figure Style - whitegrid

Changing the style to whitegrid will add a gray grid in the background, making it easier to determine specific values of the point plot. 

```python
sns.set_style('whitegrid')

sns.catplot(x='age', y='masculinity_important', data=masculinity_data, hue = 'feel_masculine', kind='point', order=['18 - 34', '35 - 64', '65 and up'])
plt.show()
```

#### Other Styles

The other styles are variants on the white and white grid:
* Ticks is similar to white, but adds small tick marks to the x and y axes. 
* Dark provides a gray background to the white style.
* Dark grid provides a gray background with white grid lines.

### Changing the Palette

Seaborn's .set_pallette() function allows for changes to the color of the main elements of the plot. There are preset palettes provided by Seaborn or a custom palette can be created. 

#### Diverging Palettes

Seaborn's group of preset palettes are called diverging palettes. These palettes are great when the visualization has two ends of the scale that are opposites and there is a neutral midpoint. Below are two examples of diverging palette options, Red-Blue, Purple-Green. If an underscore r is added to the palette name, it reverses the palette's colors. 



```python
sns.set_palette("RdBu")

category_order=['Not at all', 'Not Very', 'Somewhat', 'Very']
sns.catplot(x='how_masculine', data = masculinity_data, kind = 'count', order = category_order)
plt.show()
```

#### Sequential Palettes

Another pre-defined palette option set are sequential palettes where the colors move from light to dark values. Sequential palettes are best used when visualizing a variable on a continuous scale. 



```python
mpg = sns.load_dataset('mpg')
sns.set_style('white')
sns.set_palette('PuRd')
sns.relplot(x='horsepower', y='mpg', data=mpg, kind='scatter', hue='cylinders', size='cylinders')
```

#### Custom Palettes

Custom palettes can be created by passing Seaborn's .set_palette() function a list of custom colors or hex color codes. 

#### Change the Scale

Using Seaborn's .set_context() function allows for the customization of the scale of the plot elements and labels. The out of the box scale options, from smallest to largest are: paper, notebook, talk, poster. The default context is paper. A larger scale like talk are useful for posters or presentations.

### Adding Titles and Labels: Part 1

Plot titles and informative axes labels ensure that the audience is able to comprehend the visualization. Seaborn plots create two different types of objects: FacetGrid and AxesSubplot. To determine which type of object your plot is, assign the plot to a variable, typically named g, and check the variable type. The subplot assigned to g is an AxesSubplot. 

A FacetGrid consists of one or more axes subplots. 

#### FacetGrid vs. AxesSubplot Objects

.relplot() and .catplot() both support making subplots which means that they create FacetGrid objects when they are called. In contrast, single type plot functions like .scatterplot() and .countplot() return a single axes subplot object.




```python
g = sns.scatterplot(x='height', y='weight', data=df)
type(g)
```

### Adding Titles and Labels: Part 2

To set a title for an AxesSubplot object, the plot is assigned to a variable, g, and g.set_title() function is provided a text object with the plot's title. The placement of the title can be customized using they y = keyword argument. 

```python
#FacetGrid

g = sns.catplot(x="Region", y='Birthrate',data=gdp_data, kind='box')
g.fig.suptitle("New Title", y=1.03)
```

```python
# AxesSubplot

g=sns.boxplot(x="Region", y='Birthrate',data=gdp_data)
g.set_title('New Title', y=1.03)
```

#### Titles for Subplots

If we divide the GDP data into groups and use the col = parameter to create subplots for each group type, we can set the title for the graph as a whole using the g.fig.suptitle() function and each subplot can be given a title using g.set_titles(). Included {col_name} will automatically fill in the column name that the subplots are based on. 

```python
g = sns.catplot(x="Region", y='Birthrate',data=gdp_data, kind='box', col='Group')

g.fig.suptitle('New Title', y=1.03)
g.set_titles('This is {col_name}', y=1.03)
```

#### Adding Axis Labels

Again, create a graph and assign it to a variable. Then call the .set() function and set the parameters xlabel= and ylabel= with the axis names.

```python
g = sns.catplot(x='Region', y='Birthrate', data=gdp_data, kind = 'box')
g.set(xlabel='New X Label', ylabel='New Y Label')

plt.show()
```

#### Rotating x-Axis Tick Labels

Sometimes, as seen in this GDP data, the x-axis tick labels overlap. In this case, the matplotlib .xticks() function is called and the parameter rotation = is set to 90.

```python
g = sns.catplot(x='Region', y='Birthrate', data=gdp_data, kind = 'box')
plt.xticks(rotation=90)
plt.show()
```
