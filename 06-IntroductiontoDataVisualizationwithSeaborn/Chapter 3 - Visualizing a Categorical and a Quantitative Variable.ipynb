## Chapter 3 - Visualizing a Categorical and a Quantitative Variable

Categorical variables are made up of typically a small number of possible values known as categories. Plots like count plots and bar plots are commonly used when making comparisons between different groups. Count plots displays the number of observations in each category. Similar to relplot to show relational plots, Seaborn has a .catplot() function to create different types of categorical plots. Like .relplot(), .catplot() can create subplots using col = and row = keyword arguments.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

masculinity_data = pd.read_csv('C:\\datacamp\\06-IntroductiontoDataVisualizationwithSeaborn\\data\\masculinity.csv')

sns.catplot(x='how_masculine', data=masculinity_data, kind='count')
plt.show()
```

#### Changing the Order

Like .relplot(), the order in which the responses appear can be customized by passing a list of the preferred order to the keyword argument order = .

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

masculinity_data = pd.read_csv('C:\\datacamp\\06-IntroductiontoDataVisualizationwithSeaborn\\data\\masculinity.csv')

category_order = ['Not at all', 'Not Very', 'Somewhat', 'Very']
sns.catplot(x='how_masculine', data=masculinity_data, kind='count', order=category_order)
plt.show()
```

### Bar Plots

Bar plots show a mean of a quantitative variable per category. The bar plot below uses the tips dataset and show the average tip on each day of the week. Notice that Seaborn automatically shows 95% confidence intervals for the means. The confidence interval means that assuming the data is a random sample, we can be confident that 95% of the values in each group lies within the confidence shown by the thin black bar extending beyond the colored bar. The confidence interval can be disabled by setting the ci= keyword argument to None. 

```python
tips = sns.load_dataset('tips')
tips.columns
sns.catplot(x='day', y='total_bill', data = tips, kind='bar' )
plt.show()
```

```python
tips = sns.load_dataset('tips')
tips.columns
sns.catplot(x='day', y='total_bill', data = tips, kind='bar', ci=None )
plt.show()
```

#### Changing the Orientation of a Bar Plot

By switching the x and y parameters, the bar chart goes from vertical to horizontal. 

```python
tips = sns.load_dataset('tips')
tips.columns
sns.catplot(y='day', x='total_bill', data = tips, kind='bar' )
plt.show()
```

### Creating a Box Plot

A box plot shows the distribution of quantitative data. The colored box represents the 25th to 75th percentile and the line in the middle of the box represents the median. The whiskers, extending vertically above and below the colored box documents the distribution and the floating point represent outliers. Box plots are commonly used to prepare the distribution of a quantitative variable across different groups of a categorical variable.

```python
sns.catplot(x='day', y='total_bill', data = tips, kind = 'box')
```

#### Change the Order of Categories

Like the other plots covered, box plots can use a custom order to show the boxes, based on the list provided to the keyword argument order = .

```python
sns.catplot(x='time', y='total_bill', data = tips, kind = 'box')
```

```python
sns.catplot(x='time', y='total_bill', data = tips, kind = 'box', order =['Dinner', 'Lunch'])
```

#### Omitting Outliers

To omit the outliers, the keyword argument sym = is set to empty string. sym= can also be used to change the appearance of the outliers. 

```python
sns.catplot(x='time', y='total_bill', data = tips, kind = 'box', sym="")
```

#### Changing the Whiskers

By default the whiskers extend to 1.5 * the interquartile range (IQR), the 25th = 75th percentile. The whis = keyword argument can be used to modify the IQR, whis = 2.0 will modify whisker range to 2 x IQR. Alternately, the whiskers be defined with specific upper and lower percentiles by passing whis = a list of the lower and upper values. For example, passing whis = [5,95] will have the lower whisker being drawn at the 5th percentile and the upper whisker at the 95th percentile. To have the whiskers visualize the min and max values, pass whis = the list [0,100]. When the entire range of the data is shown by whiskers visualzing the min and the max values, there will be no outliers.

### Point Plots

Point plots show the mean of a quantitative variable for the observations of each category plotted as a single point. The vertical bars extending above and below the points show the 95% confidence intervals, or the level of uncertainty about these means.

#### Point Plots vs Line Plots

Both show the mean of a quantitative variable as well as the 95% confidence interval for the mean. But line plots are relational plots, so both the x and the y axis are quantitative variables, usually time on the x-axis. Point plots have a categorical variable, usually along the x-axis.

#### Point Plots vs Bar Plots

It may appear that point plots show the same information as bar plots since both plot types show the mean of a quantitative variable and the 95% confidence intervals for the mean. But this example demonstrates the difference between point and bar plots. The bar plot shows the percentage of men who think it is important that others see them as masculine by subgroup of whether they see themselves as masculine. The point plot below shows the same information, but it's easier to compare the heights of the subgroup points when they are stacked above each other. In the point plot, it's also easier to look at the distances in slope between the categories than it is to compare the heights of the bars.

```python
sns.catplot(x='age', y='masculinity_important', data=masculinity_data, kind='bar', hue='feel_masculine')
```

```python
sns.catplot(x='age', y='masculinity_important', data=masculinity_data, kind='point', hue='feel_masculine', order = ['18 - 34', '35 - 64', '65 and up'])
```

#### Disconnecting the Points

Sometimes it may be better to remove the lines connecting each point when you only wish to compare within a category group and not between them. To do this, use the join = keyword argument and set it to False.

```python
sns.catplot(x='age', y='masculinity_important', data=masculinity_data, kind='point', hue='feel_masculine', join = False, order = ['18 - 34', '35 - 64', '65 and up'])
```

#### Displaying the Median

To have the points and confidence interval to be calculated from the median rather than the mean, import the median function from the NumPy library and use the estimator = keyword argument and set it to median. Median is more robust to outliers, so if the data has a lot of outliers, the median may be a better statistic to use. 

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point')
```

```python
from numpy import median
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point', estimator=median)
```

#### Customing the Confidence Intervals

To add caps to the confidence intervals, the horizontal lines across the top and bottom of the confidence intervals, use the capsize = keyword argument, giving it a value of the desired width of the caps. 

#### Turning off the Confidence Intervals

Like the other plot options, using the ci= parameter and setting it to None will remove the confidence interval bars from the chart. 

```python
sns.catplot(x='smoker', y='total_bill', data=tips, kind='point', capsize = .2)
```
